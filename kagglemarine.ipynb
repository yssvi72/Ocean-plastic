```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.image as mpimg

# Read the image
img = mpimg.imread(r'C:/Users/Yashyashsvi/Desktop/demo/underwater_plastics/runs/detect/valid/images/val_batch0.jpg')

# Display the image with detected objects
plt.figure(figsize=(10, 8))
plt.imshow(img)
plt.axis('off')

# Read the labels file
labels_file = r'C:/Users/Yashyashsvi/Desktop/demo/underwater_plastics/runs/detect/train/labels/val_batch0.txt'
with open(labels_file, 'r') as f:
    lines = f.readlines()

# Define class names
class_names = ['Mask', 'can', 'cellphone', 'electronics', 'gbottle', 'glove', 'metal', 'misc', 'net', 'pbag', 'pbottle', 'plastic', 'rod', 'sunglasses', 'tire']

# Iterate through each line in the labels file
for line in lines:
    # Split the line and extract information
    class_index, x_center, y_center, width, height = map(float, line.split())
    class_index = int(class_index)
    
    # Convert normalized coordinates to pixel coordinates
    img_height, img_width, _ = img.shape
    xmin = int((x_center - width / 2) * img_width)
    ymin = int((y_center - height / 2) * img_height)
    xmax = int((x_center + width / 2) * img_width)
    ymax = int((y_center + height / 2) * img_height)
    
    # Get the class name corresponding to the class index
    class_name = class_names[class_index]
    
    # Create a rectangle patch
    rect = patches.Rectangle((xmin, ymin), xmax - xmin, ymax - ymin, linewidth=1, edgecolor='r', facecolor='none')
    plt.gca().add_patch(rect)
    
    # Add the object name as text
    plt.text(xmin, ymin - 5, class_name, color='r', fontsize=8, fontweight='bold')

# Save the image with annotations
annotated_image_path = r'C:/Users/Yashyashsvi/Desktop/demo/underwater_plastics/runs/detect/train/labels/val_batch0_annotated.jpg'
plt.savefig(annotated_image_path, bbox_inches='tight')

# Display the image in the notebook
plt.show()
```
